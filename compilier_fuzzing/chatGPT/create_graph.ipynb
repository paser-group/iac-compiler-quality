```python
import matplotlib.pyplot as plt
import pandas as pd
size = df['bert_score'].size
#result = [score for score in np.concatenate(df['bert_score']) if score != torch.tensor(0)]
result = np.array([score for score in np.concatenate(df['bert_score']) if score != 0]).reshape(size,size-1)

# Plot the histogram
plt.hist(result, bins=5, edgecolor='black', alpha=0.7, label=np.arange(0,len(result)+1,1))

# Label the x-axis and y-axis
plt.xlabel('BERT F1 Score')
plt.ylabel('Frequency')

# Set the tick marks to only show whole numbers
plt.gca().set_xticks(np.linspace(0.7825, 0.914, 5))
plt.gca().set_yticks(np.arange(0, 7, 1))

# Show the plot
plt.show()
```
